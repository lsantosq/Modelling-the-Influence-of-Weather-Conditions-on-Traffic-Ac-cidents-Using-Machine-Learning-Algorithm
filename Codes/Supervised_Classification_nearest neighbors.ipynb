```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.markers
from matplotlib.colors import ListedColormap
from mpl_toolkits import mplot3d
import pandas as pd
from sklearn import datasets,neighbors
from sklearn.neighbors import NearestNeighbors
from sklearn.tree import DecisionTreeClassifier
```

```python
#importing data file
incidents=pd.read_csv('Master_Data_clustering_and_regression.csv',encoding='latin1')
```

```python
#Supervised Learning- Clustering using KN neighbors
Acc_tot=incidents['Total number of incidents']
Acc_tot_Class=incidents['Class_Total_inc']
temp=incidents['Mean Temp (Â°C)']
rain=incidents['Total Rain (mm)']
snow=incidents['Total Snow (cm)']
wind=incidents['Spd of Max Gust (km/h)']

#data consilidation
data_tot_acc=np.zeros((948,4))
target_tot_acc=np.zeros(948)

count=0

while count<948:
    data_tot_acc[count,0]=temp[count]
    data_tot_acc[count,1]=rain[count]
    data_tot_acc[count,2]=snow[count]
    data_tot_acc[count,3]=wind[count]
    target_tot_acc[count]=Acc_tot_Class[count]
    count=count+1

#training and testing sets
data_train=data_tot_acc[0:948:2,:]
data_test=data_tot_acc[1:948:2,:]
target_train=target_tot_acc[0:948:2]
target_test=target_tot_acc[1:948:2]
```

```python
#Clustering using all attributes
X_data_train=data_train
X_data_test=data_test

# finding optimum number of neighbors-uniform weights
k_all=np.zeros(200)
Accuracy_all=np.zeros(200)
count=0
k=1

while k<201:
    
    clf=neighbors.KNeighborsClassifier(k, weights='uniform')
    clf.fit(X_data_train,target_train)
    Accuracy=clf.score(X_data_test,target_test)
    k_all[count]=k
    Accuracy_all[count]=Accuracy
    k=k+1
    count=count+1

fig, ax=plt.subplots()
plt.grid(linewidth=0.25)
Scatter, =ax.plot(k_all,Accuracy_all*100,'o',label="Accuracy - Uniform Weights", color='red')
ax.set_xlabel('number of neighbors')
ax.set_ylabel('Acuuracy, %')
ax.legend()
plt.show()
print()


# finding optimum number of neighbors-distance weights
k_all=np.zeros(200)
Accuracy_all=np.zeros(200)
count=0
k=1
while k<201:
    
    clf=neighbors.KNeighborsClassifier(k, weights='distance')
    clf.fit(X_data_train,target_train)
    Accuracy=clf.score(X_data_test,target_test)
    k_all[count]=k
    Accuracy_all[count]=Accuracy
    k=k+1
    count=count+1

fig, ax=plt.subplots()
plt.grid(linewidth=0.25)
Scatter, =ax.plot(k_all,Accuracy_all*100,'o',label="Accuracy - Distance Weights", color='blue')
ax.set_xlabel('number of neighbors')
ax.set_ylabel('Acuuracy, %')
ax.legend()
plt.show()
```
